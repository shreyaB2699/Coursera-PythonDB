```python
%sql select country, avg(test_score) as avg_test_score_country_wise from INTERNATIONAL_STUDENT_TEST_SCORES group by country
```

```python
# converting the results of queries into data frames

"""
This can be done by saving the result set into a variable and then assigning it as dataframe
"""

var = %sql select country, avg(test_score) as avg_test_score_country_wise from INTERNATIONAL_STUDENT_TEST_SCORES group by country
df_var = var.DataFrame()
print(df_var)
type(df_var)
```
