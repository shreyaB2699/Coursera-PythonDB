```python
#importing pandas and creating dataframe fromthe csv file

import pandas
df_chicago = pandas.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
```

```python
# Changing from dataframe to table

%sql --persist df_chicago
```

```python
# Querying

%sql select * from df_chicago
```

```python
# printing rows

%sql select count(*) as No_of_Rows from df_chicago
```

```python
# Community areas in chicago having hardship index greater than 50.0

%sql select count(community_area_name) as no_of_comm_area from df_chicago where hardship_index > 50.0
```

```python
# MAximum value of hardship index

%sql select max(hardship_index) as max_hardship_index_val from df_chicago
```

```python
# Community area having highest hardship index

%sql select community_area_name from df_chicago where hardship_index = (select max(hardship_index) from df_chicago)
```

```python
#community areas where per-capita income is greater than 60000

%sql select community_area_name from df_chicago where per_capita_income_ > '60000'
```

```python
#scatter plot using the variables per_capita_income_ and hardship_index

plot_var = %sql select per_capita_income_, hardship_index from df_chicago
plot_var = plot_var.DataFrame()
#print(plot_var)
#type(plot_var)
import matplotlib.pyplot as plt
plt.scatter(plot_var.per_capita_income_, plot_var.hardship_index)

# We can see there is a negative correlation between these two variables
```
