##### Data is loaded into the database using db2 load operations in the GUI of ibm cloud database console. We start by writing queries and exploring data as follows:

```python
#Problem 1 - total number of crimes recorded in the CRIME table

%sql select count(*) as total_number_of_crimes from CHICAGO_CRIME_DATA
```

```python
#Problem 2 - Retrieve first 10 rows from the CRIME table

%sql select * from CHICAGO_CRIME_DATA LIMIT 10
```

```python
# Problem 3 - How many crimes involve an arrest?

%sql select count(*) as crimes_involving_arrest from CHICAGO_CRIME_DATA where arrest = 'TRUE'
```

```python
# Problem 4 - Which unique types of crimes have been recorded at GAS STATION locations?

%sql select distinct(primary_type) as distinct_crimes_at_gas_station from CHICAGO_CRIME_DATA where location_description = 'GAS STATION'
```

```python
# Problem 5 - In the CENSUS_DATA table list all Community Areas whose names start with the letter 'B'

%sql select COMMUNITY_AREA_NAME as comm_area_name_starting_from_B from CENSUS_DATA where COMMUNITY_AREA_NAME like 'B%'
```

```python
# Problem 6 - Which schools in Community Areas 10 to 15 are healthy school certified?

%sql select NAME_OF_SCHOOL from CHICAGO_PUBLIC_SCHOOLS where COMMUNITY_AREA_NUMBER between '10' and '15' and HEALTHY_SCHOOL_CERTIFIED = 'Yes'
```

```python
#Problem 7 - What is the average school Safety Score?

%sql select avg(SAFETY_SCORE) as average_safety_score from CHICAGO_PUBLIC_SCHOOLS
```

```python
# Problem 8 - List the top 5 Community Areas by average College Enrollment [number of students]

%sql select NAME_OF_SCHOOL, avg(COLLEGE_ENROLLMENT) as average_college_enrollment from CHICAGO_PUBLIC_SCHOOLS group by NAME_OF_SCHOOL 
```

```python
# Problem 9 - Use a sub-query to determine which Community Area has the least value for school Safety Score

%sql select COMMUNITY_AREA_NAME from CHICAGO_PUBLIC_SCHOOLS where SAFETY_SCORE = (select min(SAFETY_SCORE) from CHICAGO_PUBLIC_SCHOOLS)
```

```python
# Problem 10 - [Without using an explicit JOIN operator] Find the Per Capita Income of the Community Area which has a school Safety Score of 1

%sql select PER_CAPITA_INCOME from CENSUS_DATA where COMMUNITY_AREA_NUMBER = (select COMMUNITY_AREA_NUMBER from CHICAGO_PUBLIC_SCHOOLS where SAFETY_SCORE = '1')
```
